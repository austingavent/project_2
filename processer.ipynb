```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
from datetime import datetime, timedelta
%matplotlib inline
import yfinance as yf
import mibian as mb
```

```python
source_apple = Path("Resources/apple_df.csv")
source_amd = Path("Resources/amd_df.csv")
source_msft = Path("Resources/msft_df.csv")
source_tsla = Path("Resources/tsla_df.csv")
tsla_df = pd.read_csv(source_tsla)
apple_df = pd.read_csv(source_apple)
amd_df = pd.read_csv(source_amd)
msft_df = pd.read_csv(source_msft)
```

```python
amd_df = amd_df.drop(columns = ['Unnamed: 0'])
tsla_df = tsla_df.drop(columns = ['Unnamed: 0'])
apple_df = apple_df.drop(columns = ['Unnamed: 0'])
msft_df = msft_df.drop(columns = ['Unnamed: 0'])
```

```python
#Set Time Delta and % out columns
xmsft_df = msft_df
xmsft_df['ExpirationDate'] =  pd.to_datetime(msft_df['ExpirationDate'], infer_datetime_format=True)
xmsft_df['DataDate'] =  pd.to_datetime(msft_df['DataDate'], infer_datetime_format=True)
xmsft_df['time delta'] = xmsft_df['ExpirationDate'] - xmsft_df['DataDate']
xmsft_df['% out'] = xmsft_df['StrikePrice'] / xmsft_df['UnderlyingPrice'] - 1

#Slice Test Data
msft_test = msft_df.loc[msft_df['time delta'] < np.timedelta64(21, 'D')]
msft_test = msft_test.loc[(msft_test['% out'] > -.3) & (msft_test['% out'] < .3)]

#Setup expiration value column by testing the expiration
msft_test['exp?'] = msft_test['ExpirationDate'] == msft_test['DataDate']
msft_test['xp'] = msft_test.loc[msft_test['exp?'] == True, 'AskPrice']


#Rename Columns
msft_test = msft_test[['OptionKey', 'DataDate', 'PutCall', 'ExpirationDate', 'UnderlyingPrice', 'AskPrice', 'BidPrice', 'StrikePrice', 'time delta', '% out', 'exp?']]
msft_test.to_csv('msft_test.csv')
```

```python
#Calculate volume by date
mf = msft_df
mf['Total Volume'] = mf.groupby(['DataDate']).Volume.transform('sum')
mf['Call Volume'] = mf.loc[mf['PutCall'] == 'call', 'Volume']
mf['Put Volume'] = mf.loc[mf['PutCall'] == 'put', 'Volume']
msft_volume = apf.groupby('DataDate').agg({'Call Volume': 'sum', 'Put Volume' : 'sum'})
msft_volume['Total Volume'] = msft_volume.loc[:,['Call Volume','Put Volume']].sum(axis=1)

#Create put signals
msft_volume['Percent of cPut'] = msft_volume['Put Volume']/msft_volume['Total Volume']
msft_volume['Put signals'] = msft_volume['Percent of cPut'] > .6

#Create Call volume
msft_volume['Percent of calls'] = msft_volume['Call Volume']/msft_volume['Total Volume']
msft_volume['Call signals'] = msft_volume['Percent of calls'] > .6
msft_volume.to_csv('msft_volume.csv')
```
