```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline
```

```python
investment = 100_000
shares = investment / aapl_df['Close'].iloc[0]
shares
```

```python
file_path = Path("../personal project/aapl_2012-2020.csv")
aapl_df = pd.read_csv(file_path)
aapl_df.tail(10)
```

```python
#Go biweekly
aapl = aapl_df.loc[::10, :]
```

```python
aapl['pct_change'] = aapl['Close'].pct_change()
aapl['bi-returns'] =  aapl['pct_change'] * investment
aapl['portfolio'] = aapl['Close'] * shares
aapl['returns'] = aapl['portfolio'] - investment
aapl['call_premium'] = aapl['Close'] * .006
aapl
```

```python
aapl.to_csv('basic_aapl.csv')
```
