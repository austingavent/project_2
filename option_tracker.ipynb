```python
#source_master = Path("../Resources/choice_df.csv")
source_apple = Path("Resources/master_apple.csv")
source_amd = Path("Resources/master_amd.csv")
source_msft = Path("Resources/master_msft.csv")
source_tsla = Path("Resources/master_tesla.csv")

#master_df = pd.read_csv(source_master)
tsla_df = pd.read_csv(source_tsla)
apple_df = pd.read_csv(source_apple)
amd_df = pd.read_csv(source_amd)
msft_df = pd.read_csv(source_msft)
```

```python
amd_df['ExpirationDate'] =  pd.to_datetime(amd_df['ExpirationDate'], infer_datetime_format=True)
amd_df['DataDate'] =  pd.to_datetime(amd_df['DataDate'], infer_datetime_format=True)
amd_df['time delta'] = amd_df['ExpirationDate'] - amd_df['DataDate']
amd_df['% out'] = amd_df['StrikePrice'] / amd_df['UnderlyingPrice'] - 1

amd_df.tail()
```

```python
df = pd.DataFrame(amd_df)
df = df.set_index(pd.to_datetime(df.DataDate, infer_datetime_format=True))
df['OptionKey'] = df['OptionKey'].map(lambda x:str(x)[:-10])
df = df.drop(columns = ['Unnamed: 0', 'AskSize', 'BidSize'])

df.head()
```

```python
#track PnL code- filter for only puts, slice dataframe length and sort
df = df[(df['ExpirationDate'] <= '2019-12-30') & (df.PutCall == 'put')]
df = df.sort_values(['OptionKey','time delta'])
                                                   
df.head()
```

```python
#drop all strikes with volume and open interest < 10, time delta > 1 week

amd_df = amd_df[amd_df["Volume"]>10]
amd_df = amd_df[amd_df["time delta"]<= "7 days"]
amd_df
```

```python
#calculate a 20 day rolling standard deviation of option volume and create column to record strikes over this SD
rolling_sd = 20 

amd_df['vol sd'] = amd_df['Volume'].rolling(window = rolling_sd).std()/2 + amd_df['Volume'].rolling(window = rolling_sd).mean()

#create options volume signal 
amd_df['options vol signal'] = np.where(amd_df['Volume'] > amd_df['vol sd'], 1.0, 0.0)

        
amd_df
```
